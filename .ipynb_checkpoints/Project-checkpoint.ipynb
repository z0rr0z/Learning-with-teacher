**Задачи (этапы) исследования:**
1. Провести загрузку и общий обзор данных.

7. Сформировать итоговые заключения и рекомендации заказчику.

### Описание данных 

Файл market_file.csv - данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении:
- id — номер покупателя в корпоративной базе данных;
- Покупательская активность — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень»;
- Тип сервиса — уровень сервиса, например «премиум» и «стандарт»;
- Разрешить сообщать — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель;
- Маркет_актив_6_мес — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента;
- Маркет_актив_тек_мес — количество маркетинговых коммуникаций в текущем месяце;
- Длительность — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте;
- Акционные_покупки — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев;
- Популярная_категория — самая популярная категория товаров у покупателя за последние 6 месяцев;
- Средний_просмотр_категорий_за_визит — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца;
- Неоплаченные_продукты_штук_квартал — общее число неоплаченных товаров в корзине за последние 3 месяца;
- Ошибка_сервиса — число сбоев, которые коснулись покупателя во время посещения сайта;
- Страниц_за_визит — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.

Файл market_money.csv - данные о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом:
- id — номер покупателя в корпоративной базе данных;
- Период — название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц';
- Выручка — сумма выручки за период;

Файл market_time.csv - данные о времени (в минутах), которое покупатель провёл на сайте в течение периода:
- id — номер покупателя в корпоративной базе данных;
- Период — название периода, во время которого зафиксировано общее время;
- минут — значение времени, проведённого на сайте, в минутах.

Файл money.csv - данные о среднемесячной прибыли покупателя за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю:
- id — номер покупателя в корпоративной базе данных;
- Прибыль — значение прибыли.

### Описательная статистика и графики числовых параметров

Определим функцию для построения графиков числовых параметров.

```python
# Функция строит "сетку" графиков для различных столбцов датафрейма.
# Параметры функции:
# df - датафрейм с исходными данными;
# columns - список столбцов, по которым строятся графики;
# num_cols - число столбцов "сетке" графиков;
# title - общий заголовок для "сетки";
# hue - столбец, для уникальных значений которого строятся дополнительные графики

def plot_quant(df, columns, num_cols, title, hue=None):
    num_rows = -(-len(columns) // num_cols) # округление вверх

    fig = plt.figure(figsize=(num_cols*5, num_rows*3))
    outer_grid = fig.add_gridspec(num_rows, num_cols, hspace=0.5, wspace=0.2)

    i = 0

    for column in columns:
        inner_grid = outer_grid[i].subgridspec(2, 1, height_ratios=[2, 1], 
                                               hspace=0, wspace=0)
        ax = inner_grid.subplots(sharex=True)
        sns.histplot(data=df, x=column, stat='percent', common_norm=False, bins=20, 
                     legend=False, ax=ax[0], label='общая совокупность')\
        .set(ylabel='Частотность, %')
        ax[0].set_title(column, fontsize=12)
        ax2 = ax[0].twinx()
        
        if hue:
            sns.kdeplot(data=df, x=column, common_norm=False, hue=hue, fill=True, cut=0,
                        legend=(not i), ax=ax2)
        
        sns.boxplot(data=df, x=column, legend=False, ax=ax[1])\
                    .set(xlabel='Значение')
        ax2.set(ylabel='', yticks=[])
        
        # получаем параметры легенды первого графика и убираем ее
        if hue and not i: 
            lines, labels = ax[0].get_legend_handles_labels()
            legend = ax2.get_legend()
            handles = lines + legend.legend_handles
            texts = labels + [hue + ': ' + x.get_text() for x in legend.get_texts()]
            legend.remove()

        i += 1

    # размещаем общую легенду для сетки     
    if hue:
        fig.legend(handles, texts, loc='upper center', 
                   bbox_to_anchor=(0.5, -0.1), fontsize=12)

    fig.suptitle(title, va='bottom', size=16, y=1.1)
    plt.subplots_adjust(
        left=0, 
        right=1, 
        top=1, 
        bottom=0, 
        wspace=0, 
        hspace=0
    )
    plt.show()
```

Выведем описательные статистики и построим графики для числовых параметров во всех датасетах, в графиках акже рассмотрим различия распределений параметров зависимости от целевого признака (покупательская активность).

```python
market_file.describe()
```

Достаточно значимое различие в зависимости от целевого признака мы видим по следующим параметрам:

### Графики категориальных параметров

Определим функцию для построения графиков категориальных параметров.

```python
def plot_qual(df, columns, num_cols, title, hue=None):
    num_rows = -(-len(columns) // num_cols) # округление вверх
    fig = plt.figure(figsize=(num_cols*8, num_rows*5), layout='constrained')

    i = 0

    for column in columns:
        ax = fig.add_subplot(num_rows, num_cols, i+1)
        sns.countplot(data=df, y=column, stat='percent', legend=(not i), ax=ax, 
                      color='grey',  edgecolor='black', 
                      order=df[column].value_counts().index, label='общая совокупность') \
                     .set(xlabel='Частотность, %', ylabel='', title=column)
        if hue:
            sns.histplot(data=df, y=column, stat='percent', legend=(not i), 
                         common_norm=False, ax=ax, hue=hue, multiple='dodge', 
                         shrink=.5, alpha=0.5)
        ax.set(ylabel='')
        
        if hue and not i:
            lines, labels = ax.get_legend_handles_labels()
            legend = ax.get_legend()
            handles = lines + legend.legend_handles
            texts = labels + [hue + ': ' + x.get_text() for x in legend.get_texts()]
            legend.remove()
            
        i += 1

    if hue:
        fig.legend(handles, texts, loc='upper center', 
                   bbox_to_anchor=(0.5, -0.1), fontsize=12)
    
    fig.set_constrained_layout_pads(wspace=0.1, hspace=0.1)
    fig.suptitle(title, va='bottom', size=16)
    plt.show()
```

Построим графики для категориальных параметров в датасете market_file с выделением групп в зависимости от значений целевого признака.

```python
plot_qual(
    df=market_file, 
    columns=market_file.select_dtypes(include='object').columns, 
    num_cols=2, 
    title='Датафрейм: market_file',
    hue='покупательская_активность'
)
```

Мы видим, что бо́льшая часть признаков имеет неодинаковую долю классов, в том числе и целевой признак - покупательская активность. Это необоходимо будет учесть при выборе целевой метрики для моделей.

Также необходимо отметить, что клиенты с разной покупательской активностью отличаются по параметру "популярная_категория", в частности наиболее сильно - по доле категории "Мелкая бытовая техника и электроника". Учтем это при построении моделей.

##  Построение и анализ моделей

### Создание пайплайнов

Заменим целевой признак на бинарный (значение "снизилась" - класс 1).

```python
label_mapping = {'Снизилась': 1, 'Прежний уровень': 0}
label_encoder = LabelEncoder()
market_combined['покупательская_активность'] = label_encoder.fit_transform(
    market_combined['покупательская_активность'].map(label_mapping)
) 
```

Разделим данные на тренировочные и тестовые. Учитывая дисбаланс в целевом признаке, сделаем стратифицированную выборку. Кроме того, исключим признаки "выручка_текущий_месяц", "время_текущий_месяц" и "маркет_актив_тек_мес" (см. выше).

```python
market_combined_cut = market_combined.drop(
    ['выручка_текущий_месяц', 'время_текущий_месяц', 'маркет_актив_тек_мес'], 
    axis=1
).copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    market_combined_cut.drop('покупательская_активность', axis=1),
    market_combined_cut['покупательская_активность'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE,
    stratify = market_combined_cut['покупательская_активность'])
```

Разделим признаки на 3 группы: качественные бинарные, качественные множественные и количественные

```python
ohe_columns = X_train.select_dtypes(include='object').columns.to_list()
num_columns = X_train.select_dtypes(include='number').columns.to_list()
```

Создадим пайплайн для кодирования качественных бинарных признаков.

```python
ohe_pipe = Pipeline(
    [('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
    ]
)
```

Создадим пайплайн для отбора и масштабирования количественных признаков.

```python
num_pipe = Pipeline(
    [('scaler', MinMaxScaler()),
     ('selector', SelectKBest(k='all'))
    ]
)
```

Создадим общий пайплайн подготовки данных.

```python
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('num', num_pipe, num_columns),
    ],
    remainder='passthrough'
)
```

Создадим итоговый пайплайн с подготовкой данных и моделью.

```python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', LogisticRegression())
])
```

Создадим словарь для гиперпараметров моделей и параметров пайплайнов:
- модели: DecisionTreeClassifier, LogisticRegression, SVC;
- гиперпараметры для DecisionTreeClassifier: макс. глубина - от 2 до 4, мин. число объектов в узле - от 2 до 4, мин. число объектов в листе - от 1 до 2;
- гиперпараметры для LogisticRegression: тип регуляризации - l1 или l2, сила регуляризации - от 1 до 5;
- гиперпараметры для SVC с полиномиальным ядром: степень полинома - от 1 до 2;
- для каждой модели (кроме дерева решений) переберем два варианта масштабирования числовых признаков: StandardScaler и MinMaxScaler;
- количество лучших числовых признаков для SelectKBest - от 5 до 10, для логистической регрессии - от 3 до 10. 

```python
param_grid = [

    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 5),
        'models__min_samples_split': range(2, 5),
        'models__min_samples_leaf': range(1, 3),
        'preprocessor__num__scaler':  ['passthrough'],
        'preprocessor__num__selector__k': range(5, 11)
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE,
            solver='liblinear',
        )],
        'models__penalty': ['l1', 'l2'],
        'models__C': range(1, 6),
        'preprocessor__num__scaler':  [StandardScaler(), MinMaxScaler()],
        'preprocessor__num__selector__k': range(3, 11)
    },

     # словарь для модели SVC()
    {
        'models': [SVC(kernel='poly', random_state=RANDOM_STATE)],
        'models__degree': range(1, 3),
        'preprocessor__num__scaler':  [StandardScaler(), MinMaxScaler()],
        'preprocessor__num__selector__k': range(5, 11)
    }
]
```

### Выбор метрики и подбор гипераметров моделей

Поскольку целью является удержание активности, самой важной характеристикой модели является ее способность верно предсказывать клиентов, активность которых снизится. Т.е. важна минимизация ошибки 2 рода (False Negative), когда мы неверно определяем клиента как того, у кого активность сохранится на прежнем уровне. Поскольку данные несбалансированы, для данной цели подходит метрика recall.

Однако ошибки 1 рода (False Positive) тоже важны (хотя и в меньшей степени, чем ошибки 2 рода): если мы ошибочно посчитаем клиента как того, у кого активность снизится, мы потеряем деньги на избыточные мероприятия по его удержанию, которые, возможно, и не требовались. Для данной цели подходит метрика precision.

Таким образом, нам необоходимо использовать метрику, учитывающую как recall, так и precision, но отдающую больший вес recall. Для этого подходит метрика F-beta с параметром beta > 1. Установим его равным 2.

```python
f2_score = make_scorer(fbeta_score, beta=2)
```

Запустим сплошной подбор параметров с помощью GridSearchCV. При этом наряду с целевой метрикой рассчитаем также собственно recall и ROC AUC.

```python
grid_search = GridSearchCV(
    pipe_final,
    param_grid,
    cv=5,
    scoring={'roc_auc': 'roc_auc', 'recall': 'recall', 'f2': f2_score},
    refit='f2',
    n_jobs=-1,
)
```

```python
%%time
grid_search.fit(X_train, y_train)
```

Лучшая модель - логистическая регрессия. Выведем параметры и значение метрики для данной модели.

```python
grid_search.best_params_
```

Лучшей моделью является логичтическая регрессия с l2-регуляризацией и "обратной силой" 3, с использованием стандартизации и 9 числовых факторов из 10 исходных.

Посмотрим, какие факторы попали в модель.

```python
list(grid_search.best_estimator_.named_steps['preprocessor'].get_feature_names_out())
```

Из исходных отброшен признак "выручка_предыдущий_месяц".

```python
print(f'Метрика f2 для лучшей модели на тренировочных данных: {grid_search.best_score_:.3f}')
```

Рассмотрим 10 "моделей-лидеров".

```python
pd.DataFrame(grid_search.cv_results_).sort_values(['rank_test_f2', 'rank_test_recall', 'rank_test_roc_auc']).head(10)
```

Лидеры по целевой метрике - логистические регрессии.

Значение метрики достаточное, чтобы говорить о приемлемом качестве модели.

##  Сегментация покупателей

Выделим сегмент покупателей исходя их двух параметров: имеют высокую вероятность (более 0.8) снижения покупательской активности, принесли заказчику высокую (выше медианы) прибыль.
По сути это клиенты, снижение активности (или даже потеря) которых может плохо повлиять на финансовое положение заказчика, в связи с чем именно на них нужно в первую очередь концентрировать мероприятия по удержанию.
Для сравнения выделим также контрсегмент с такой же прибылью (выше медианы), но с низкой (менее 0.2) вероятностью снижения покупательской активности.

Найдем вероятности снижения активности для всех покупателей.

```python
X = market_combined_cut.drop(['покупательская_активность'], axis=1).copy()
y = market_combined_cut['покупательская_активность']

y_proba = model.predict_proba(
    grid_search.best_estimator_.named_steps['preprocessor'].transform(X)
)
```

Добавим в объединенный датесет данные по прибыли, а также расчетные вероятности снижения активности.

```python
clients = market_combined.join(money, how='left')
clients = clients.drop('покупательская_активность', axis=1)
clients['предсказ_вероят'] = y_proba[:, 1]
```

Найдем медиану прибыли.

```python
profit_median = clients['прибыль'].median()
```

Добавим новый столбец с типом сегмента.

```python
def segment(row):
    if row['прибыль'] > profit_median and row['предсказ_вероят'] > 0.8:
        return 'целевой'
    elif row['прибыль'] > profit_median and row['предсказ_вероят'] < 0.2:
        return 'контр'
    else:
        return np.nan

clients['сегмент'] = clients.apply(segment, axis=1)
```

Возможно, именно этим обусловлено то, что в целевом сегменте наблюдается последовательное увеличение выручки.

Таким образом, можно выделить следующие особенности целевого сегмента:
- невысокая активность поведения на сайте;
- последовательное увеличение выручки, в т.ч. за счет роста выручки от клиентов с наиболее популярной категорией "Косметика и аксессуары";
- высокая доля акционных покупок;
- высокая доля товаров в категориях "Товары для детей" и "Косметика и аксессуары";
- невысокая маркетинговая активность.
Из этого можно сделать вывод, что целевая группа, скорее всего, более ориентирована на повседневные товары, которые стремится покупать по акциям.
Можно предположить, что это, наряду с небольшим количеством маркетинговых мероприятий, не способствует формированию лояльности к заказчику.
Данный факт, в свою очередь, при некоторых обстоятельствах (например, закончились акции, у конкурентов появились чуть более выгодные предложения) может приводить к снижению активности целевой группы.

Для сегмента покупателей с высокой прибыльностью и высокой вероятностью снижения активности можно предложить следующие мероприятия:
- предусмотреть персонализацию сайта, что может способствовать посещению большего числа страниц;
- предусмотреть персонализацию выдач товаров (возможно, стоит предлагать товары, по которым выручка в целевом сегменте снижается - например, категории "Техника для красоты и здоровья");
- предусмотреть возможность выдачи готовых "наборов" товаров исходя из предпочтений (это может способствовать увеличению времени нахождения на сайте и числа просмотров категорий, а также снижению числа неоплаченных покупок);
- чтобы мотивировать покупателей приобретать "наборы", можно предусмотреть по ним скидки, снизив объемы нетаргетированных акций;
- увеличить число маркетинговых коммуникаций (рассылки, звонки и т.д.);
- рассмотреть возможность мониторинга предложений конкурентов с тем, чтобы при формировании персональных предложений предусматривать чуть более выгодные варианты.
